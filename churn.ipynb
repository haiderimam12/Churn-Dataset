```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
ch = pd.read_csv('churn.csv')
ch.head(7)
```

Steps involved in model building
1. problem statement ---- come to know the target variable
2. data gathering
3. data cleaning / data pre-processing
4. sampling divide data into train and test ---- train data and build the model and test data to do the testing
* mode data s in train (80%) and test(20%)
* testing should not be done on same data on which model was build
* sampling should be random
5. fit the model ---- build / train the model
6. do the prediction and evaluate the model ---- basically check the performane of the model

1. problem statement ---- predict the status
2. data gathering ---- data already provided
3. data cleaning ---- remove the nulls 

```python
ch.isnull().sum()[ch.isnull().sum() > 0]
# it will find the null values and give the count of null values
```

* labels (non-numeric data)
* model should be build only on numbers
* ordinal data should be converted using replace and nominal labelencoder should be used

```python
ch_select_dtype = ch.select_dtypes(include = ['object'])
ch_select_dtype.head(5)
```

```python
ch.info()
```

```python
ch.Contract.replace({'Month-to-month': 0, 'One year': 1, 'Two year': 2}, inplace = True)
ch.head(10)
```

```python
ch.drop(['customerID'], axis = 1)
```

```python
le.fit_transform(ch.TotalCharges)
```

```python
ch.TotalCharges
```

```python
ch.TotalCharges = ch.TotalCharges.replace({'':np.nan})
# converting space to nulls
```

```python
ch.TotalCharges.isnull().sum()
```

```python
le = LabelEncoder()
```

```python
ch[ch.select_dtypes(include='object').columns] = ch[ch.select_dtypes(include='object').columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
# train_test_split  spilts the data into train and test
```

```python
ch_train, ch_test = train_test_split(ch, test_size = 0.2)
ch_train_x = ch_train.iloc[:,0:-1]
ch_train_y = ch_train.iloc[:,-1]
ch_test_x = ch_test.iloc[:,0:-1]
ch_test_y = ch_test.iloc[:,-1]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(ch_train_x,ch_train_y)
predictions=lr.predict(ch_test_x)
tab=confusion_matrix(ch_test_y,predictions)
print(tab)
print("------------------------------------------------------")
print(classification_report(ch_test_y,predictions))
```

```python
from sklearn.tree  import DecisionTreeClassifier
lr = DecisionTreeClassifier()
lr.fit(ch_train_x,ch_train_y)
predictions=lr.predict(ch_test_x)
tab=confusion_matrix(ch_test_y,predictions)
print(tab)
print("------------------------------------------------------")
print(classification_report(ch_test_y,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
lr = RandomForestClassifier(class_weight = 'balanced')
lr.fit(ch_train_x,ch_train_y)
predictions=lr.predict(ch_test_x)
tab=confusion_matrix(ch_test_y,predictions)
print(tab)
print("------------------------------------------------------")
print(classification_report(ch_test_y,predictions))
```
